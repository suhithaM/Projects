```python
#Import pandas module
import pandas as pd
```

```python
#Load the dataset into a dataframe
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv")
```

```python
df.head(5) #to get the overview of the dataframe
```

```python
#Finding number of duplicated rows in the dataframe
duplicate_rows = df.duplicated().sum()
print("Number of duplicated rows:", duplicate_rows)
```

```python
#Remove duplicated rows from the dataframe
df.drop_duplicates(inplace=True)
df.shape   #check the shape of dataframe after removing duplicates
```

```python
#Finding missing values for all columns
missing_values = df.isnull().sum()
print("Missing values in each columns:", missing_values)
```

```python
#Find how many missing values are in 'WorkLoc' column
missing_workloc = df['WorkLoc'].isnull().sum()
print("Number of missing values in WorkLoc column:", missing_workloc)
```

```python
#Find the value counts for the column WorkLoc
df['WorkLoc'].value_counts()
```

```python
#Identify the value that is most frequent (majority) in the WorkLoc column
print("mode_WorkLoc:", df['WorkLoc'].mode())
```

```python
#Impute (replace) all the empty rows in the column WorkLoc with the value that was identified as majority
df['WorkLoc'].fillna('Office', inplace = True)
```

```python
#Verify if imputing was successful
df['WorkLoc'].isna().sum()
```

```python
#List out the various categories in the column 'CompFreq'
print (df['CompFreq'].value_counts())  
```

Create a new column named 'NormalizedAnnualCompensation'
If the CompFreq is Yearly then use the exising value in CompTotal
If the CompFreq is Monthly then multiply the value in CompTotal with 12 (months in an year)
If the CompFreq is Weekly then multiply the value in CompTotal with 52 (weeks in an year)

```python
norm_annual_comp = []

for i in range(len(df)):
    
    if df['CompFreq'].iloc[i] == 'Weekly':
        norm_annual_comp.append(df['CompTotal'].iloc[i]*52)
    elif df['CompFreq'].iloc[i] == 'Monthly':
        norm_annual_comp.append(df['CompTotal'].iloc[i]*12)
    else:
        norm_annual_comp.append(df['CompTotal'].iloc[i]*1)

df['NormalizedAnnualCompensation']= norm_annual_comp
df.head()
```
